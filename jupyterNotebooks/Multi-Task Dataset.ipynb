```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
import sys
sys.path.insert(0, '/home/mm22d016/Multi-Task')
from multiTask_dataset import MultiTask_Dataset
from torch.utils.data import DataLoader
from PolymerSmilesTokenization import PolymerSmilesTokenizer
```

```python
export = pd.read_csv('../data/export.csv')
export
```

```python
unique_smiles = export['smiles'].unique()
unique_data = {'smiles': unique_smiles}
for prop in ['Eea', 'Egb', 'Egc', 'Ei']:
    prop_values = []
    for smile in unique_smiles:
        value = export[(export['smiles'] == smile) & (export['property'] == prop)]['value'].values
        if len(value) > 0:
            prop_values.append(value[0])
        else:
            prop_values.append(None)
    
    unique_data[prop] = prop_values
```

```python
unique_df = pd.DataFrame(unique_data)
unique_df = unique_df.fillna(-99)
unique_df
```

```python
unique_df.to_csv('../data/multiTask.csv',index=False)
```
