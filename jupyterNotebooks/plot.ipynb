```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Your data
categories = ['Original', 'ST(Best)', 'ST_Aug(Best)', 'MT(Best)', 'MT_Aug(Best)', 'ST(Avg.)', 'ST_Aug(Avg.)', 'MT(Avg.)', 'MT_Aug(Avg.)']

rmse_values = [0.44, 0.457936871, 0.502858044, 0.495880435, 0.503610247, 0.462211009, 0.508091518, 0.473424061, 0.513662295]

sdrmse_values = [0, 0.010738092, 0.01665268, 0.057447103, 0.050134793, 0.003028135, 0.004831129, 0.012611791, 0.011141437]


r2_values = [0.92, 0.913901174, 0.896145296, 0.933342473, 0.894857381, 0.912345223, 0.893963392, 0.914535444, 0.89078583]

sdr2_values = [0, 0.00618495, 0.008798977, 0.018933848, 0.021744018, 0.001132965, 0.002051275, 0.009675104, 0.004835992]


```

```python
bar_width = 0.5
index = np.arange(len(categories))
fig, ax = plt.subplots(figsize=(10, 6))


ax.bar(index, rmse_values, bar_width, label='RMSE')
#ax.bar(index, r2_values, bar_width, label='R2')

ax.errorbar(index, rmse_values, yerr=sdrmse_values, fmt='o', capsize=5, label='SDRMSE', color='black', markerfacecolor='white')
#ax.errorbar(index, r2_values, yerr=sdr2_values, fmt='o', capsize=5, label='SDR2', color='black', markerfacecolor='white')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Egc_RMSE')

# Set x-axis limits to zoom in
ax.set_xlim(-0.5, len(categories) - 0.5)

ax.set_xticks(index)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.yaxis.set_major_locator(plt.MaxNLocator(nbins='auto'))
ax.legend()

plt.subplots_adjust(top=0.85)

plt.savefig('Egc_RMSE.png', bbox_inches='tight')
plt.show()
```
